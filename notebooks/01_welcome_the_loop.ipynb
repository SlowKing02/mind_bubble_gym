## Welcome, Coach

You are training a **robot player**. It cannot see the future. It only tries moves, gets **points**, and slowly gets better.

That is **reinforcement learning**:

1. **State** — what the game looks like now
2. **Action** — a move you allow (left, right, …)
3. **Reward** — points (good or bad)
4. **Policy** — the robot's strategy

![RL loop](../docs/assets/rl_loop.svg)


## Experiment: zero training

First run **random** moves on **CartPole** (balance a pole on a cart). Expect chaos.

```python
import gymnasium as gym
import numpy as np

env = gym.make('CartPole-v1')
state, info = env.reset()
steps = 0
for _ in range(200):
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)
    steps += 1
    if terminated or truncated:
        break
print(f'Random agent survived {steps} steps. Trained agents reach 195+.')
env.close()
```

## What you should notice

- Short runs at first — the pole falls fast.
- Same game every time, different luck.

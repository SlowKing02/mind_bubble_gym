## Capstone: reward design

Production teams forget this; middle schoolers get it fast:

| Reward you write | Agent learns |
|------------------|--------------|
| +1 per step pole stays up | Balance forever, ignore finishing |
| +100 only for moving right | Wrong behavior, fast |
| Passenger dropoff only | Ignores pickup — bad trips |

**Same lesson in 2026:** eval metrics for LLMs, recommender clicks, fraud rules — **you get what you measure**.
